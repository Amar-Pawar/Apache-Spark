```python
import tweepy
import matplotlib.pyplot as plt
from  textblob import TextBlob 
import csv
import time
import os
from dotenv import load_dotenv
load_dotenv('.env')
```

```python
#authentication keys to access twitter API
consumer_key=os.getenv("CONSUMER_KEY")
consumer_secret=os.getenv("CONSUMER_SECRET")
access_token =os.getenv("ACCESS_TOKEN")
access_secret=os.getenv("ACCESS_SECRET")
```

```python
# connected to jump server of twitter
auth = tweepy.OAuthHandler(consumer_key,consumer_secret)
```

```python
# now we can connect from jump server to web server of twitter
auth.set_access_token(access_token,access_secret)
```

```python
# now we can connect to API storge server of twitter
api_connect=tweepy.API(auth)
```

```python
# topic to fetch data from twitter
tweet_data=api_connect.search('football')
```

```python

pos="positive"
neg="negative"
pos_count =0
neg_count =0
opinion = ["pos","neg","pos_count","neg_count"]
# opinion=["sentiments","count"]
with open('data.csv', 'w') as csv_file:
      csv_writer = csv.DictWriter(csv_file, fieldnames=opinion)
      csv_writer.writeheader()

while True:
# printing line by line
   with open('data.csv', 'a') as csv_file:
      for tweet in tweet_data:
         print(tweet.text)
         analysis=TextBlob(tweet.text) # here it will apply NLP\
         print(analysis.sentiment)
         # now checking polarity only
         if analysis.sentiment.polarity > 0:
            print("positive")
            pos_count=pos_count+1
         elif analysis.sentiment[0]<0:
            print("Negative")
            neg_count =neg_count+1
         else :
            print("Neutral")
         csv_writer = csv.DictWriter(csv_file, fieldnames=opinion)
         info = {
                  "pos": pos,
                  "neg": neg,
                  "pos_count":pos_count,
                  "neg_count":neg_count
                  }
         csv_writer.writerow(info)
         print(pos_count,neg_count)
         time.sleep(1)   
```
